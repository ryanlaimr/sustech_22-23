# Visualization

python提供了matplotlib这个包，允许我们以比较美观的形式将我们的数据可视化

matplotlib基于numpy实现，因此他们通常一起被import



最简单的方法 plot()画点（或者画线，默认画线）

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,2,3,4])
plt.show()
```

Matplotlib既支持面向对象的绘图语法，也支持命令式语法。

命令式语法imperative syntax(有时称为“状态机”语法)发出一串命令，所有这些命令都作用于最近的图形或轴。

面向对象语法object-oriented syntax显式地作用于感兴趣的对象(图形、轴等)。

```python
# 命令式
import matplotlib.pyplot as plt
drug_conc = [0, 0.5, 1, 2, 3, 4]
response = [0, 0.2, 0.5, 0.8, 0.9, 0.95]
plt.plot(drug_conc, response)
plt.title("Drug response")
plt.xlabel("Drug concentration")
plt.ylabel("Response")
plt.show()
```

```python
# 面向对象
import matplotlib.pyplot as plt
drug_conc = [0, 0.5, 1, 2, 3, 4]
response = [0, 0.2, 0.5, 0.8, 0.9, 0.95]
fig = plt.figure()
ax = fig.add_axes([0.5,0,1,1]) # 左、底、宽、高
ax.plot(drug_conc, response)
ax.set_title("Drug response")
ax.set_xlabel("Drug concentration")
ax.set_ylabel("Response")
plt.show()

fig2 = plt.figure()
ax2 = fig2.add_axes([5,0,1.5,1.5]) # 左、底、宽、高
ax2.plot(drug_conc, response)
ax2.set_title("Drug response")
ax2.set_xlabel("Drug concentration")
ax2.set_ylabel("Response")
plt.show()
```

首先创建figure对象 plt.figure( figsize=None,…)

然后设置axes add_axes (rect)



当然我们可以创建一个包含有多张图的figure

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].set_ylabel("Response")
ax[0, 1].plot(drug_conc, response)
ax[1, 0].plot(drug_conc, response)
ax[1, 0].set_xlabel("Drug concentration")
ax[1, 0].set_ylabel("Response")
ax[1, 1].set_xlabel("Drug concentration")
plt.show()
```

Errorbar 误差条图

```python
import matplotlib.pyplot as plt

drug_conc = [0, 0.5, 1, 2, 3, 4]
response = [0, 0.2, 0.5, 0.8, 0.9, 0.95]
std = [0, 0.1, 0.2, 0.1, 0.1, 0.2]
fig, ax = plt.subplots(nrows=1, ncols=1) # 画一张图
ax.errorbar(drug_conc, response, yerr = std, fmt = 'ro--', ecolor='black', capsize=5) # 画errorbar
ax.set_title('drug response', fontsize=15, color='red', y=1.0, loc='center') # 设置标题
ax.set_xlabel('dose') # x轴
ax.set_ylabel('response', fontsize=15, color='blue') # y轴
plt.show() # 显示图像
```

Scatter 散点图

```python
import numpy as np

drug_conc = [0, 0.5, 1, 2, 3, 4]
response1 = [0, 0.2, 0.5, 0.8, 0.9, 0.95]
response2 = [0, 0.1, 0.2, 0.4, 0.8, .9]
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.scatter(drug_conc, response1, s=np.array(response1)*200, c='r', marker='o', alpha=0.5) # 散点
ax.scatter(drug_conc, response2, s=np.array(response2)*200, c='b', marker='o', alpha=0.5)
ax.set_title('drug response', fontsize=15, color='red', y=1.0, loc='center')
ax.set_xlabel('dose')
ax.set_ylabel('response', fontsize=15, color='blue')
plt.show()
```

GridSpec：自定义子图的摆放

```python
# Define sinc function
def s(x):
    a = np.where(x==0., 1., np.sin(x)/x)
    return a

x = np.arange(0., 10., 0.1)
y = np.exp(x)

t = np.linspace(-15., 15., 150)
z = s(t)

fig = plt.figure(figsize=(9, 7))

# subplot: linear plot of exponential
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, y, 'b')
ax1.set_xlim(0, 10)
ax1.set_xlabel('time (ms)')
ax1.set_ylabel('distance (mm)')
ax1.set_title('exponential')
# subplot: semi-log plot of exponential
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, y, 'g')
ax2.set_yscale('log')
# ax2.semilogy(x, y, 'g')
ax2.set_xlim(0, 10)
ax2.set_xlabel('time (ms)')
ax2.set_ylabel('distance (mm)')
ax2.set_title('exponential')
# subplot: wide subplot of sinc function
ax3 = fig.add_subplot(2, 1, 2)
# ax3 = fig.add_subplot(2, 2, (3, 4))
ax3.plot(t, z, 'r')
ax3.axhline(color='gray', zorder=-1)
ax3.axvline(color='gray', zorder=-1)
ax3.set_xlim(-15, 15)
ax3.set_xlabel('angle (deg)')
ax3.set_ylabel('electric field')
ax3.set_title('sinc function')

# show
fig.tight_layout()
plt.show()
```

小格的选取也可以像列表的切片一样

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 3)
ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('gs[0, :]')
ax2 = fig.add_subplot(gs[1, :-1])
ax2.set_title('gs[1, :-1]')
ax3 = fig.add_subplot(gs[1:, -1])
ax3.set_title('gs[1:, -1]')
ax4 = fig.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')
ax5 = fig.add_subplot(gs[-1, -2])
ax5.set_title('gs[-1, -2]')
plt.show()
```

也可以在创建了格子后手动添加格子

```python
fig = plt.figure(constrained_layout=True)
gs0 = fig.add_gridspec(1, 2)

gs00 = gs0[0].subgridspec(2, 3)
gs01 = gs0[1].subgridspec(3, 2)

for a in range(2):
    for b in range(3):
        fig.add_subplot(gs00[a, b])
        fig.add_subplot(gs01[b, a])
```

常用方式

```python
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up the axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)
# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical', color='gray')
x_hist.invert_yaxis()
y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```

```python
num1 = '123'
num2 = '99'
num1 = num1[::-1]
num2 = num2[::-1]
nums = []
for i in range(len(num1)):
    num = []
    for j in range(len(num2)):
        num.append(int(num1[i])*int(num2[j]))
    nums.append(num)
ans = []
max_len = len(num1) + len(num2)
for i in range(max_len):
    sum = 0
    for j in range(len(nums)):
        if(j>i):
            break
        if(len(nums[j])>i-j):
           sum += nums[j][i-j]
    ans.append(sum)
for i in range(len(ans)-1):
    ans[i+1] += ans[i] // 10
    ans[i] = ans[i]%10
if(ans[-1]==0):
    ans = ans[:-1]
ans = ans[::-1]
for i in range(len(ans)):
    print(ans[i],end='')
```
